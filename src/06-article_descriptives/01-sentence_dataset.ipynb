```python
from pathlib import Path

import numpy as np
import pandas as pd
import sqlalchemy as sa
from sqlalchemy.orm import Query
from sqlalchemy.orm import joinedload

import src.db.models.bert_data as bm
import src.db.models.open_discourse as od
from src.db.connect import make_engine
```

```python
out_path = Path("/home/lukas/overleaf/bert_populism/tables")
```

```python
engine = make_engine("DB")
```

```python
def add_midrules(latex: str) -> str:
    new = []
    for line in latex.splitlines():
        if line.startswith(r"\multirow"):
            new.append(r"\midrule")
        if line.startswith("sum"):
            new.append(r"\midrule")
        new.append(line)
    return "\n".join(new)
```

# How many speeches / sentences?


```python
query = (
    Query(od.Speech)
    .join(bm.Sample)
    .join(od.Faction)
    .filter(
        od.Speech.electoral_term.in_([18, 19]), od.Faction.id != -1, od.Speech.politician_id != -1
    )
    .group_by(
        od.Speech.electoral_term,
        od.Speech.session,
        od.Faction.abbreviation,
        od.Speech.politician_id,
    )
    .with_entities(
        od.Speech.electoral_term,
        od.Faction.abbreviation,
        sa.func.max(bm.Sample.sentence_no).label("n_sents"),
    )
)

with engine.connect() as conn:
    df_sents = pd.read_sql(query.statement, conn)

df_sents.columns = ["Term", "Party", "sentences"]
```

```python
df_sents.head()
```

```python
table = (
    df_sents.groupby(["Term", "Party"])["sentences"]
    .agg(func=[len, sum, np.mean, np.std])
    .rename(
        {
            "len": "Speeches",
            "mean": "Avg. sentences",
            "std": "Std. sentences",
            "sum": "Sentences",
        },
        axis=1,
    )
    .sort_values(["Term", "Speeches"], ascending=[True, False])
).reset_index()

table = pd.concat(
    [
        table,
        pd.DataFrame(
            [["sum", "", table["Speeches"].sum(), table["Sentences"].sum(), "", ""]],
            columns=table.columns,
        ),
    ]
).set_index(["Term", "Party"])
```

```python
table
```

```python
tex = (
    table.style.set_table_styles(
        [
            {"selector": "toprule", "props": ":toprule;"},
            {"selector": "bottomrule", "props": ":bottomrule;"},
        ]
    )
    .format(thousands=",", precision=3)
    .to_latex()
)


lines = tex.splitlines()
new = []
for i, line in enumerate(lines, 1):
    #    print(i, line)
    line = line.replace(r"\multirow[c]", r"\multirow[t]")
    if i == 3:
        line = r"Term & Party & Speeches & Sentences & Avg. sentences & Std. sentences \\\midrule"
    if i == 4:
        continue

    if i == len(lines) - 3:
        line = line + r"\midrule"

    new.append(line)

tex = "\n".join(new)

(out_path / "n_dataset.tex").write_text(tex)

print(tex)
```

```python
speeches = s.query(bm.Sample).join(od.Speech).with_entities(od.Speech.id).distinct().subquery()

query = (
    s.query(speeches)
    .join(od.Speech, speeches.c.id == od.Speech.id, isouter=True)
    .join(od.Faction, isouter=True)
    .group_by(od.Speech.electoral_term, od.Faction.abbreviation)
    .with_entities(od.Speech.electoral_term, od.Faction.abbreviation, sa.func.count())
)

df_speeches = pd.read_sql(query.statement, engine)

df_speeches = df_speeches.sort_values(["electoral_term", "abbreviation"])
df_speeches.columns = ["Term", "Party", "speeches"]
```

```python
df = pd.merge(df_speeches, df_sents, on=["Term", "Party"], how="inner")
```

```python
df = pd.concat(
    [df, pd.DataFrame([["sum", "", df.speeches.sum(), df.sentences.sum()]], columns=df.columns)],
    ignore_index=True,
)
df = df.set_index(["Term", "Party"])
```

```python
style = df.style.set_table_styles(
    [
        {"selector": "toprule", "props": ":toprule;"},
        {"selector": "bottomrule", "props": ":bottomrule;"},
    ]
).format(subset=["sentences", "speeches"], thousands=",")

latex = style.to_latex(
    multirow_align="t",
    # clines="skip-last;index",
)

latex = add_midrules(latex)

_ = (out_path / "n_dataset.tex").write_text(latex)
```

# How many sentences labeled?


```python
samples = (
    s.query(bm.Sample)
    .options(joinedload(bm.Sample.raw_labels))
    .join(bm.Label)
    .join(od.Speech)
    .join(od.Faction)
    .filter(bm.Sample.used_in_batch != None)
    .with_entities(od.Speech.electoral_term, od.Faction.abbreviation, bm.Sample)
)

rows = []
for term, faction, sample in samples:
    for label in sample.raw_labels:
        row = (
            term,
            faction,
            sample.id,
            label.unsure,
            label.pop_antielite,
            label.pop_pplcentr,
            label.souv_eliteless,
            label.souv_pplmore,
            label.ideol_left,
            label.ideol_right,
        )
        rows.append(row)

df = pd.DataFrame(
    rows,
    columns=[
        "Term",
        "Party",
        "sample_id",
        "unsure",
        "antielite",
        "pplcentr",
        "eliteless",
        "pplmore",
        "left",
        "right",
    ],
)
```

```python
cols = [
    "antielite",
    "pplcentr",
    "eliteless",
    "pplmore",
    "left",
    "right",
]

pivot = pd.pivot_table(
    df,
    index=["Term", "Party", "unsure"],
    values=cols,
    aggfunc="sum",
)
```

```python
pivot = pivot[cols]
pivot = pivot.reset_index()

sums = pivot[cols].sum(axis=0).tolist()


pivot = pd.concat(
    [pivot, pd.DataFrame([["sum", "", "", *sums]], columns=pivot.columns)],
    ignore_index=True,
)
pivot = pivot.set_index(["Term", "Party", "unsure"])
```

```python
style = pivot.style.set_table_styles(
    [
        {"selector": "toprule", "props": ":toprule;"},
        {"selector": "bottomrule", "props": ":bottomrule;"},
    ]
).format(subset=cols, thousands=",")

latex = style.to_latex(
    multirow_align="t",
)

latex = add_midrules(latex)

_ = (out_path / "n_labels.tex").write_text(latex)
```
